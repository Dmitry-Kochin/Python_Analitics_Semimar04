1.2 Изучите стоимости недвижимости


```python
houses = pd.read_csv('kc_house_data.csv', sep=',')
houses.head()
```

```python
houses['price'] = houses['price'].fillna(0).astype(np.int64)
houses.head()
```

```python
plt.figure(figsize=(8,8))
sns.lineplot(x=houses['sqft_living'], y=houses['price'])
plt.title('Зависимость стоимости домов от Жилой Плошади')
plt.xlabel('ЖилаяПлошадь')
plt.ylabel('Стоимость Дома')
```

Вывод - стоимость дома имеет восходящий тренд при увеличении жилой площади

```python
pd.options.display.float_format = '{:,.0f}'.format
houses['price'].describe().round(2)
```

```python
pd.options.display.float_format = '{:,.0f}'.format
houses['price'].describe().round(2)
sns.histplot(houses["price"]) 
```

```python
plt.figure(figsize=(8,8))
sns.histplot(x=houses['price'], bins=15)
plt.title('Распределение Стоимости Домов')
plt.xlabel('Стоимость Дома')
plt.ylabel('Количество Домов')
```
